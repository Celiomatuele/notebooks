This notebook demonstrates the functionality of CPT-DL - an API for downloading data from the IRI Data Library

Import the library:

```python
import xarray as xr 
import cptdl as dl
import datetime as dt 
```

The concept is that the IRI Data Library uses an in-URL query language called ingrid - meaning that any subset of data can be downloaded by typing the right set of characters into a web browser. 

CPT-DL stores a set of template-URLS, similar to how the PyCPT v1's 'dictionary file' did. Those templates are stored in three top-level objects within CPTDL: 

```python
dl.hindcasts
```

```python
dl.forecasts
```

```python
dl.observations
```

You can see in each of those template urls, there are sections delimited by curly-braces { and }. Those are python format-strings. you can see how they work next:

```python
x = 99999
print( f"I have {x} dogs")
```

luckily, we don't have to do that all manually. CPT-DL does that automatically for us, and all we have to do is specify enough arguments to denominate the forecast data we want. 



```python
# first, we choose a template to download 
template = dl.hindcasts['CanSIPSv2.PRCP']

# then we decide where to save the data to
destination_file = "cansips_prcp.tsv" 

# then we specify all of our forecast arguments - these hould be familiar from PyCPT v1
kwargs = { 
  'fdate': dt.datetime.now(),
  'first_year': 1982, 
  'final_year': 2018, 
  'predictor_extent': {
    'east': 90,
    'west': 0, 
    'north': 90, 
    'south': 0
  }, 
  'lead_low': 1.5,
  'lead_high': 3.5, 
  'target': 'Jul-Sep',
  'filetype': 'cptv10.tsv'
}

# this returns an xarray dataarray: 
da = dl.download(template, destination_file, use_dlauth=False, verbose=True, **kwargs)  
```

```python
da.prec.mean('T').plot()
```

You can list all of the GCMS or Observations easily using python dictionary methods (since dl.hindcasts, dl.forecasts, and dl.observations are all python dictionaries) 

```python
print([key for key in dl.observations.keys() ])
```

If you want to delimit the search, add a boolean at the end of the list comprehension: 

```python
print([key for key in dl.observations.keys() if 'ENACTS' in key ])
```
