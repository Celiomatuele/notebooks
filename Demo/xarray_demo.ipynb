#### Import Xarray
here we are using cptcore to load some sample gridded data- its NMME output over South Asia

```python
import xarray as xr 
import datetime as dt 
import cptcore as cc 
X, Y = cc.load_southasia_nmme()
```

Lets inspect X:

```python
X
```

From this, you can see that X is a 3-Dimensional array, labelled with T, Y, and X dimensions and coordinates corresponding to each one. Ti, Tf, and S correspond to the T dimension, and represent "start of target period", "end of target period", and "initialization" respectively. 

Now, lets select a given year and plot the data

```python
X_1980 = X.sel(T=dt.datetime(1980, 7, 31, 12))
```

```python
X_1980
```

Now that we have selected a single point in time, `X_1980` is a 2D data array, and we can plot it simply: 

```python
X_1980.plot()
```

Next, lets try selecting a single latitude/longitude point, and plotting a time series, all in one line: 

```python
X.sel(X=80, Y=20).plot()
```

Other important xarray functionality includes reading and writing netcdf files:

```python
X.mean('T').to_netcdf('X_climatology.nc')
```

```python
X2 = xr.open_dataset('X_climatology.nc')
```

```python
X2
```

Notice, however, that X2 is now and `xr.Dataset` rather than an `xr.DataArray` - this is an important distinction, as DataSets can contain multiple DataArrays, and while they share a lot of the same functionality, they don't share it all. PyCPT libraries use DataArrays - so you need to extract the 'prec' DataArray from the above:

```python
x_dataarray = X2.prec
```

Those are the most important Xarray functionalities for PyCPT users - more information is found [here](https://docs.xarray.dev/en/stable/)
