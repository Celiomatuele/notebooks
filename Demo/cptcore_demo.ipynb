This notebook demonstrates the usage of CPT-Core for data analytics

```python
import cptcore as cc 
```

CPT-Core analysis takes the form of a set of python functions, which accept and return xarray data arrays. The following cells demonstrate a subset of the CPT-Core functions:

This one implements Canonical Correlation Analysis 

```python
x, y = cc.load_southasia_nmme()
f = x.isel(T=slice(-1, None))
x = x.isel(T=slice(None,-1))
hindcasts, forecasts, skills, patterns_x, patterns_y = cc.canonical_correlation_analysis(x, y, F=f)
```

```python
patterns_x
```

Principal Components Regression:

```python
x, y = cc.load_southasia_nmme()
f = x.isel(T=slice(-1, None))
x = x.isel(T=slice(None,-1))
hindcasts, forecasts, skills, patterns_x = cc.principal_components_regression(x, y, F=f)
```

```python
patterns_x
```

Deterministic Skill Scores:

```python
x, y = cc.load_southasia_nmme()
f = x.isel(T=slice(-1, None))
x = x.isel(T=slice(None,-1))
skills = cc.deterministic_skill(x, y)
```
